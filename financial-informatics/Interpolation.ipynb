```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
%matplotlib notebook
x= [0, 0.25, 1.0, 2.0, 3.0, 5.0, 7.0, 10.0, 15.0, 20.0, 30.0]
y = [0.0001, 0.008, 0.012, 0.014, 0.015, 0.02, 0.026, 0.032, 0.036, 0.0358, 0.037]
plt.plot(x, y, 'o')
plt.title("Data Points")
plt.xlabel('X-values')
plt.ylabel('Y-values');
```

```python
%matplotlib notebook
xmissing =np.linspace(0, 30, 360)
linear = interpolate.interp1d(x, y)
ylinear = linear(xmissing)
plt.plot(xmissing, ylinear, '*', label='linear')
plt.legend(loc='lower left', ncol=2)
plt.title("Linear Interpolation")
plt.xlabel('X-values')
plt.ylabel('Y-values');
```

```python
from scipy.interpolate import CubicSpline
%matplotlib notebook

cSpline = CubicSpline(x, y)
ySpline = cSpline(xmissing)
plt.plot(xmissing, ySpline, '*', label='Cubic Spline')

plt.legend(loc='lower left', ncol=2)
plt.title("Cubic Spline Interpolation")
plt.xlabel('X-values')
plt.ylabel('Y-values');
```

```python
from scipy.interpolate import CubicSpline
%matplotlib notebook
linear = interpolate.interp1d(x, y)
ylinear = linear(xmissing)
plt.plot(xmissing, ylinear, '', label='linear')

cSpline = CubicSpline(x, y)
ySpline = cSpline(xmissing)
plt.plot(xmissing, ySpline, '', label='Cubic Spline')

plt.legend(loc='lower left', ncol=2)
plt.title("Cubic Spline Interpolation")
plt.xlabel('X-values')
plt.ylabel('Y-values')
```
